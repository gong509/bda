```python
import pandas as pd
import numpy as np

df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                  [3, 4, np.nan, 1],
                  [np.nan, np.nan, np.nan, np.nan],
                  [np.nan, 3, np.nan, 4]],
                  columns=list('ABCD'))
df
```

```python
df.fillna(10)
```

```python
df
```

```python
df.fillna('missing', axis=0)
```

```python
df.fillna('missing', limit=1, axis=1)
```

```python
df.fillna('missing', limit=1)
```

```python
df.fillna('value', axis=1)
```

```python
values = {"A": 100, "B": 200, "C": 300, "D": 400}
df.fillna(value=values)
```

```python
df.fillna(value=values, limit=2)
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
display_side_by_side(df, df.fillna('missing', limit=1))
```

```python
display_side_by_side(df, df.fillna('missing', limit=2, axis=1))
```

```python
display_side_by_side(df, df.fillna('missing', limit=2))
```

```python
s= pd.Series([100, 200, 400], index=list('ACD'))
s
```

```python
df2 = pd.DataFrame(-np.ones((4, 4)), columns=list("ABCE"))
df2
```

```python
df.fillna(df2)
```

```python
display_side_by_side(df, df.fillna((df2)))
```
